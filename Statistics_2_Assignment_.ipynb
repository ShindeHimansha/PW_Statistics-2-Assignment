#1. What is hypothesis testing in statistics ?
- Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population based on sample data. It helps determine whether there is enough evidence to reject a null hypothesis in favor of an alternative hypothesis.

- Key Concepts in Hypothesis Testing:
    - Null Hypothesis (H‚ÇÄ): The default assumption that there is no effect, no difference, or no relationship in the population.
      - Example: "There is no difference in average test scores between two teaching methods."
    - Alternative Hypothesis (H‚ÇÅ or Ha): The statement we want to test against the null hypothesis. It suggests that there is an effect, difference, or relationship.
      - Example: "Students taught using Method A have higher average test scores than those taught using Method B."
    - Significance Level (Œ±): The probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 (5%) or 0.01 (1%).

    - Test Statistic: A numerical value calculated from the sample data, which helps determine whether to reject H‚ÇÄ. Examples include:
      - t-statistic (t-test)
      - z-score (z-test)
      - chi-square statistic (œá¬≤ test)
    
    - P-Value: The probability of obtaining a test result as extreme as, or more extreme than, the observed result under the assumption that H‚ÇÄ is true.
      - If p ‚â§ Œ±, reject H‚ÇÄ (strong evidence against the null hypothesis).
      - If p > Œ±, fail to reject H‚ÇÄ (not enough evidence to support Ha).
    - Type I and Type II Errors:
      - Type I Error (False Positive): Rejecting H‚ÇÄ when it is actually true.
      - Type II Error (False Negative): Failing to reject H‚ÇÄ when Ha is actually true.

#3.  What is the significance level in hypothesis testing, and why is it important?
- The significance level (Œ±) is the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true (Type I error).
- It represents the threshold for statistical significance‚Äîhow much risk we are willing to take in wrongly rejecting H‚ÇÄ.
- Common values for Œ± are 0.05 (5%), 0.01 (1%), and 0.10 (10%).

- Why is the Significance Level Important?
    - Controls Type I Error (False Positive)
        - A lower Œ± (e.g., 0.01) reduces the risk of mistakenly rejecting a true H‚ÇÄ but may increase Type II errors.
        - A higher Œ± (e.g., 0.10) increases the chance of rejecting H‚ÇÄ even if it is true.

    - Determines Decision Making in Hypothesis Testing
      - If p-value ‚â§ Œ±, reject H‚ÇÄ (strong evidence for H‚ÇÅ).
      - If p-value > Œ±, fail to reject H‚ÇÄ (not enough evidence to support H‚ÇÅ).
    
    - Balances Accuracy and Confidence
      - In critical applications (e.g., medical trials), a lower Œ± (0.01 or 0.001) is used to minimize false positives.
      - In exploratory research, a higher Œ± (0.10) may be used to allow more flexibility in detecting potential effects.

#4.  What does a P-value represent in hypothesis testing ?

#6. What are Type 1 and Type 2 errors in hypothesis testing ?
- Type I Error (False Positive)
    - Occurs when we reject the null hypothesis (H‚ÇÄ) when it is actually true.
    - It means detecting an effect when there is none.
    - Probability of making a Type I error is denoted by Œ± (significance level).
    - Example: A medical test wrongly detects a disease in a healthy person.
- Type II Error (False Negative)
    - Occurs when we fail to reject the null hypothesis (H‚ÇÄ) when it is actually false.
    - It means missing a real effect or failing to detect a true relationship.
    - Probability of making a Type II error is denoted by Œ≤.
    - Example: A medical test fails to detect a disease in a sick person.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
- One-Tailed Test
    - Tests for an effect in one specific direction (either greater than or less than).
    - The alternative hypothesis (H‚ÇÅ) is directional:
      - Right-tailed test: H‚ÇÅ: Œº > Œº‚ÇÄ (Tests if the mean is significantly greater).

    - Disadvantage: Cannot detect an effect in the opposite direction.
- Two-Tailed Test
    - Tests for an effect in either direction (greater than or less than).
    - The alternative hypothesis (H‚ÇÅ) is non-directional:
      - H‚ÇÅ: Œº ‚â† Œº‚ÇÄ (Tests if the mean is significantly different).
    - Example: A researcher tests whether a new teaching method affects exam scores (H‚ÇÅ: Œº ‚â† old score).

    - Disadvantage: Requires a larger sample size to achieve the same power as a one-tailed test.

#19. What is the F-distribution, and when is it used in hypothesis testing ?

    - F=(x1^2/d1)/(x2^2/d2)
    - where x1^2 and x2^2 are Chi-square distributed random variables,
    - and d1 and d2  are their respective degrees of freedom (df).

- The F-distribution is used in tests that involve comparing variances or multiple means. The most common applications include:
    1. ANOVA (Analysis of Variance)
    -	Used to compare the means of multiple groups.
    -	Null Hypothesis H0H_0H0: All group means are equal.
    -	Alternative Hypothesis H1H_1H1: At least one group mean is different.
    -	Example: Checking if different fertilizers affect plant growth differently.
    2. F-Test for Equality of Variances
    -	Used to test if two populations have equal variances.
    -	Null Hypothesis H0H_0H0: The variances of the two populations are equal.
    -	Alternative Hypothesis H1H_1H1: The variances are not equal.
    - Example: Checking if male and female students have the same variance in test scores.


#20. What is an ANOVA test, and what are its assumptions ?
- The ANOVA (Analysis of Variance) test is a statistical method used to determine whether there are significant differences between the means of three or more independent groups.
- ANOVA is used when:
    -	You have three or more groups and want to compare their means.
    -	Your dependent variable is continuous (e.g., test scores, heights, salaries).
    -	Your independent variable is categorical (e.g., different diets, different teaching methods).
- Hypotheses in ANOVA:
    -	Null Hypothesis (H0H_0H0): All group means are equal.
    -	Alternative Hypothesis (H1H_1H1): At least one group mean is different.
- Assumptions of ANOVA
    To ensure valid results, ANOVA relies on the following assumptions:
    1.	Independence of Observations
      -	Data in each group should be independent from each other.
    2.	Normality
      -	Each group‚Äôs data should be normally distributed.
      -	Verified using Shapiro-Wilk test or Q-Q plot.
    3.	Homogeneity of Variance (Homoscedasticity)
      -	The variance in each group should be approximately equal.
      -	Checked using Levene‚Äôs test or Bartlett‚Äôs test.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)  # Sample size
    sample_mean = np.mean(sample)  # Sample mean

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(f"Reject the null hypothesis (H0). The sample mean is significantly different from the population mean at Œ±={alpha}.")
    else:
        print(f"Fail to reject the null hypothesis (H0). No significant difference between the sample and population mean.")

# Example Usage
sample_data = [50, 52, 49, 53, 51, 48, 54, 55, 50, 52]  # Sample dataset
population_mean = 50  # Known population mean
population_std = 3  # Known population standard deviation

z_test(sample_data, population_mean, population_std)
```

```python
#2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python

import numpy as np
import scipy.stats as stats

# Step 1: Generate random sample data
np.random.seed(42)  # For reproducibility
sample_size = 30  # Number of observations
population_mean = 50  # True population mean
population_std = 5  # Population standard deviation (use for Z-test)
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)  # Simulated sample

# Step 2: Define Hypothesis Testing Function
def hypothesis_test(sample, pop_mean, pop_std=None, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation

    if pop_std:  # If population standard deviation is known, use Z-test
        z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
        test_used = "Z-test"
    else:  # If population standard deviation is unknown, use T-test
        t_score, p_value = stats.ttest_1samp(sample, pop_mean)
        test_used = "T-test"

    # Print Results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Std Dev: {sample_std:.4f}")
    print(f"{test_used} Statistic: {z_score if pop_std else t_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(f"Reject the null hypothesis (H0). The sample mean is significantly different from the population mean at Œ±={alpha}.")
    else:
        print(f"Fail to reject the null hypothesis (H0). No significant difference.")

# Step 3: Perform Hypothesis Test
hypothesis_test(sample_data, pop_mean=50, pop_std=5)  # Using Z-test
hypothesis_test(sample_data, pop_mean=50, pop_std=None)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject H0: Sample mean is significantly different from population mean at Œ±={alpha}.")
    else:
        print(f"Fail to reject H0: No significant difference from population mean.")

    z_critical = stats.norm.ppf(1 - alpha / 2)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Rejection Region")
    plt.axvline(z_score, color='blue', linestyle='dashed', linewidth=2, label=f"Z-Score: {z_score:.2f}")
    plt.axvline(-z_critical, color='red', linestyle='dashed', linewidth=2, label=f"Critical Value: {-z_critical:.2f}")
    plt.axvline(z_critical, color='red', linestyle='dashed', linewidth=2, label=f"Critical Value: {z_critical:.2f}")

    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.title("Two-Tailed Z-Test with Decision Regions")
    plt.legend()
    plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)
two_tailed_z_test(sample_data, pop_mean=50, pop_std=5)
```

```python
#5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2(pop_mean, pop_std, sample_mean, n, alpha=0.05):
    se = pop_std / np.sqrt(n)
    z_critical = stats.norm.ppf(1 - alpha / 2)

    type_1_x = np.linspace(pop_mean - 4 * se, pop_mean + 4 * se, 1000)
    type_1_y = stats.norm.pdf(type_1_x, pop_mean, se)

    type_2_x = np.linspace(sample_mean - 4 * se, sample_mean + 4 * se, 1000)
    type_2_y = stats.norm.pdf(type_2_x, sample_mean, se)

    plt.figure(figsize=(10, 5))
    plt.plot(type_1_x, type_1_y, label="Null Distribution (H0)", color='blue')
    plt.plot(type_2_x, type_2_y, label="Alternative Distribution (H1)", color='green')

    rejection_region_right = pop_mean + z_critical * se
    rejection_region_left = pop_mean - z_critical * se

    plt.fill_between(type_1_x, type_1_y, where=(type_1_x >= rejection_region_right) | (type_1_x <= rejection_region_left),
                     color='red', alpha=0.3, label="Type 1 Error (Œ±)")

    plt.fill_between(type_2_x, type_2_y, where=(type_2_x < rejection_region_right) & (type_2_x > rejection_region_left),
                     color='purple', alpha=0.3, label="Type 2 Error (Œ≤)")

    plt.axvline(pop_mean, color='blue', linestyle='dashed', label="Population Mean (H0)")
    plt.axvline(sample_mean, color='green', linestyle='dashed', label="Sample Mean (H1)")
    plt.axvline(rejection_region_right, color='red', linestyle='dashed', label=f"Critical Value (+{z_critical:.2f})")
    plt.axvline(rejection_region_left, color='red', linestyle='dashed', label=f"Critical Value (-{z_critical:.2f})")

    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Visualization of Type 1 (Œ±) and Type 2 (Œ≤) Errors")
    plt.legend()
    plt.show()

np.random.seed(42)
pop_mean = 50
pop_std = 5
sample_mean = 52
n = 30

visualize_type1_type2(pop_mean, pop_std, sample_mean, n)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject H0: The two samples have significantly different means at Œ±={alpha}.")
    else:
        print(f"Fail to reject H0: No significant difference between the sample means.")

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=52, scale=5, size=30)

independent_t_test(sample1, sample2)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_rel(sample1, sample2)

    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject H0: The paired samples have significantly different means at Œ±={alpha}.")
    else:
        print(f"Fail to reject H0: No significant difference between the paired sample means.")

    plt.figure(figsize=(8, 5))
    plt.plot(sample1, label="Sample 1", marker='o', linestyle='dashed')
    plt.plot(sample2, label="Sample 2", marker='s', linestyle='dashed')
    plt.xlabel("Sample Index")
    plt.ylabel("Values")
    plt.title("Paired Sample Comparison")
    plt.legend()
    plt.show()

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = sample1 + np.random.normal(loc=1, scale=2, size=30)

paired_t_test(sample1, sample2)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"{confidence*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
    return lower_bound, upper_bound

np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)
confidence_interval(sample)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    sample_std = np.std(sample, ddof=1)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_critical * (sample_std / np.sqrt(n))

    print(f"Margin of Error at {confidence*100}% confidence level: {moe:.4f}")
    return moe

np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)
margin_of_error(sample)
```

```python
#11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

# Example Scenario:
# A disease affects 1% of the population (Prior Probability)
# A test detects the disease 90% of the time when present (Likelihood)
# The test gives a false positive 5% of the time (False Positive Rate)
# Total Probability of Testing Positive (Evidence)

prior_prob = 0.01  # P(Disease)
likelihood = 0.9   # P(Positive Test | Disease)
false_positive = 0.05  # P(Positive Test | No Disease)
prob_no_disease = 1 - prior_prob  # P(No Disease)

evidence = (likelihood * prior_prob) + (false_positive * prob_no_disease)  # P(Positive Test)

posterior_prob = bayes_theorem(prior_prob, likelihood, evidence)

print(f"Posterior Probability (Having Disease | Positive Test): {posterior_prob:.4f}")
```

```python
#12.  Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
import scipy.stats as stats

# Sample contingency table (observed frequencies)
# Example: A survey of people‚Äôs preference for two products (A and B) based on gender
# Rows: Gender (Male, Female)
# Columns: Product Preference (A, B)

observed = np.array([[30, 20],  # Males
                     [10, 40]]) # Females

# Perform the Chi-square test for independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

# Display results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-Value: {p_value:.4f}")
print("Expected Frequencies Table:")
print(expected)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print(f"Reject H0: There is a significant association between the two categorical variables at Œ±={alpha}.")
else:
    print(f"Fail to reject H0: No significant association between the two categorical variables.")
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples

import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

f_stat = var1 / var2
df1, df2 = len(sample1) - 1, len(sample2) - 1
p_value = 2 * min(stats.f.cdf(f_stat, df1, df2), 1 - stats.f.cdf(f_stat, df1, df2))

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(f"Reject H0: The variances are significantly different.")
else:
    print(f"Fail to reject H0: No significant difference in variances.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i} Normality Test: W={stat:.4f}, p={p_value:.4f} - {'Pass' if p_value > 0.05 else 'Fail'}")

    stat, p_value = stats.levene(*groups)
    print(f"Levene‚Äôs Test for Equal Variance: W={stat:.4f}, p={p_value:.4f} - {'Pass' if p_value > 0.05 else 'Fail'}")

    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title("Boxplot for ANOVA Assumptions")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

check_anova_assumptions(group1, group2, group3)

```

```python
#24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample = np.random.normal(50, 10, 30)
sample_variance = np.var(sample, ddof=1)
pop_variance = 100

chi_square_stat = (len(sample) - 1) * sample_variance / pop_variance
p_value = stats.chi2.sf(chi_square_stat, df=len(sample) - 1)

print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: The population variance is significantly different.")
else:
    print("Fail to reject H0: No significant difference in population variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import statsmodels.api as sm

np.random.seed(42)
successes_1, nobs_1 = 50, 200
successes_2, nobs_2 = 60, 220

count = np.array([successes_1, successes_2])
nobs = np.array([nobs_1, nobs_2])

z_stat, p_value = sm.stats.proportions_ztest(count, nobs)

print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: The proportions are significantly different.")
else:
    print("Fail to reject H0: No significant difference in proportions.")
```

```python
#26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
data1 = np.random.normal(50, 10, 30)
data2 = np.random.normal(55, 15, 30)

var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)
f_stat = var1 / var2
df1, df2 = len(data1) - 1, len(data2) - 1
p_value = 2 * min(stats.f.cdf(f_stat, df1, df2), 1 - stats.f.cdf(f_stat, df1, df2))

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: The variances are significantly different.")
else:
    print("Fail to reject H0: No significant difference in variances.")

x = np.linspace(0, 5, 500)
plt.plot(x, stats.f.pdf(x, df1, df2), label=f"F-distribution df1={df1}, df2={df2}")
plt.axvline(f_stat, color='r', linestyle='--', label="Observed F-statistic")
plt.fill_between(x, stats.f.pdf(x, df1, df2), where=(x > stats.f.ppf(1-alpha/2, df1, df2)) |
                 (x < stats.f.ppf(alpha/2, df1, df2)), color='red', alpha=0.3, label="Rejection Region")
plt.title("F-Test for Variance Comparison")
plt.xlabel("F-Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
observed = np.random.randint(20, 50, size=5)
expected = np.full_like(observed, np.mean(observed))
expected = expected * (np.sum(observed) / np.sum(expected))

chi_stat, p_value = stats.chisquare(observed, expected)

print(f"Observed: {observed}")
print(f"Expected: {expected}")
print(f"Chi-Square Statistic: {chi_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: The observed distribution significantly differs from the expected distribution.")
else:
    print("Fail to reject H0: No significant difference between observed and expected distributions.")
```
